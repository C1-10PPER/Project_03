```python
from opencage.geocoder import OpenCageGeocode
import pandas as pd

# Replace with your OpenCage API key
key = "API KEY"
geocoder = OpenCageGeocode(key)

# Load your existing CSV or Excel file
df = pd.read_csv("./Raw_Data/Annual_Surface_Temp_Change.csv") 

# Clean the country names by removing extra descriptors
def clean_country_name(country):
    clean_country = country.split(",")[0]  
    return clean_country.strip() 

# Apply the cleaning function to the 'Country' column
df['Country'] = df['Country'].apply(clean_country_name)

# Define the geocoding function
def get_lat_lon(country):
    try:
        print(f"Geocoding country: {country}")
        result = geocoder.geocode(country)
        if result:
            lat = result[0]['geometry']['lat']
            lon = result[0]['geometry']['lng']
            return pd.Series([lat, lon], index=['Latitude', 'Longitude'])
        else:
            print(f"Could not find coordinates for: {country}")
            return pd.Series([None, None], index=['Latitude', 'Longitude'])
    except Exception as e:
        print(f"Error geocoding {country}: {e}")
        return pd.Series([None, None], index=['Latitude', 'Longitude']) 
# Apply the geocoding function to get lat/lon for all cleaned country names
df[['Latitude', 'Longitude']] = df['Country'].apply(get_lat_lon)

# Show the updated DataFrame with Latitude and Longitude columns
print(df[['Country', 'Latitude', 'Longitude']])

# Optionally, save the updated DataFrame back to a CSV or Excel file
df.to_csv("./Raw_Data/Annual_Surface_Temp_Change updated_file.csv", index=False) 
df
```
